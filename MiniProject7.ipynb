# • Input file contain Roll numbers and marks.
# • Create output file that add all the marks.

Screenshot (9).png

```python
def add_marks(input_file, output_file):
    total_marks = 0

    with open(input_file, 'r') as infile:
        for line in infile:
            parts = line.strip().split()
            if len(parts) >= 2:
                try:
                    mark = float(parts[1])
                    total_marks += mark
                except ValueError:
                    print(f"Invalid mark '{parts[1]}' in line: {line.strip()}")

    with open(output_file, 'w') as outfile:
        outfile.write(f"Total Marks: {total_marks}\n")

# Example usage
add_marks("input.txt", "output.txt")
```
